```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
from matplotlib.ticker import MaxNLocator
import tqdm as tqdm
```

```python
def convert_to_df(file_path):
    col = {'exp': [], 'tay': [], 'T1': [], 'T2': [], 'time': [], 'prog_time': []}
    df = pd.DataFrame(col)

    f = open(file_path, "r")

    while True:
        line = f.readline()
        if not line:
            break
        raw = line.rsplit(sep='\t')
        raw[0] = raw[0].replace("D:\\Repos\\CMC_1_Pt_Cu\\Raws\\test", '')
        raw[0] = raw[0].replace(".txt", '')
        raw[0] = raw[0].rsplit(sep='_')
        data = []
        for l in raw[0]:
            data.append(int(l.rsplit(sep='=')[1]))
        data.append(float(raw[2].replace('ms', '')))
        df.loc[len(df)] = data
    f.close()
    return df
```

```python
def get_list_of_files(file_path):
    f = open(file_path, 'r')
    files = []
    while True:
        line = f.readline()
        if not line:
            break
        files.append(line.rsplit(sep='\t')[0].replace('D:\\Repos\\CMC_1_Pt_Cu\\Raws\\test', ''))
    return files  
```

```python
def get_data(path,file,df):
    i = 0
    for file in files:
        f = open(path+file, 'r')
        line = f.readline()
        while line != '(Длина, число отсчетов)\n':
            line = f.readline()
        line = f.readline()
        j = 1
        while line != '-1\n' and j < 60:
            line = f.readline()
            if len(line.rsplit()) == 2:
                df.loc[i, str(j)] = int(line.split()[1])
                j+=1
        text = []
        while True:
            line = f.readline()
            if not line:
                break
            text.append(line)
        # df.loc[i, 'text'] = ''.join(text)
        i += 1
        f.close()
    return df
```

```python
s = r"D:\Repos\CMC_1_Pt_Cu\Raws\Result\2\run_time.txt"
df1 = convert_to_df(s)
files = get_list_of_files(s)
df1 = get_data("D:\\Repos\\CMC_1_Pt_Cu\\Raws\\Result\\2",files,df1)

s = r"D:\Repos\CMC_1_Pt_Cu\Raws\Result\3\run_time.txt"
df2 = convert_to_df(s)
files = get_list_of_files(s)
df2 = get_data("D:\\Repos\\CMC_1_Pt_Cu\\Raws\\Result\\3",files,df2)

s = r"D:\Repos\CMC_1_Pt_Cu\Raws\Result\4\run_time.txt"
df3 = convert_to_df(s)
files = get_list_of_files(s)
df3 = get_data("D:\\Repos\\CMC_1_Pt_Cu\\Raws\\Result\\4",files,df3)

path = "D:\\Repos\\CMC_1_Pt_Cu\\Raws\\Result\\4"
```
